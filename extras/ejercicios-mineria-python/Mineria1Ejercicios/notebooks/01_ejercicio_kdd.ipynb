## Insertar los imports

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
```

## Definición de la clase
A continuación se define la clase base del simulador KDD proporcionado en el ejercicio.

```python
class SimuladorKDDRetail:
    """
    Simulador educativo del proceso KDD para retail.
    Enfoque en fases iniciales: comprensión del negocio, comprensión de datos, preparación de datos.
    """
    def __init__(self):
        self.datos_transaccionales = None
        self.objetivo_analitico = None
        self.problemas_calidad = []

    def fase_1_comprension_negocio(self):
        """Simula la fase de comprensión del negocio: de problema vago a objetivo analítico claro"""
        print("="*70)
        print(" FASE 1: COMPRENSIÓN DEL NEGOCIO")
        print("="*70)
        print("\n[ESCENARIO] Reunión con gerente de retail:")
        print("Gerente: 'Necesitamos usar minería de datos para mejorar nuestras ventas.'")
        print("Analista: '¿Qué significa exactamente 'mejorar ventas'? ¿Aumentar volumen, margen, frecuencia de compra?'")
        print("Gerente: 'Queremos que los clientes compren más productos por visita.'")
        print("Analista: '¿Cuál es el objetivo cuantificable y el plazo?'")
        print("Gerente: 'Aumentar el valor promedio de ticket en un 15% en los próximos 6 meses.'")
        self.objetivo_analitico = {
            "problema_negocio": "Bajo valor promedio de ticket",
            "objetivo_analitico": "Identificar asociaciones no obvias entre categorías de productos que aumenten el valor del ticket",
            "metrica_exito": "Aumento del 15% en valor promedio de ticket",
            "plazo": "6 meses",
            "restricciones": ["Sin cambios en precios", "Sin campañas publicitarias masivas"]
        }
        print(f"\n OBJETIVO ANALÍTICO DEFINIDO:")
        for key, value in self.objetivo_analitico.items():
            print(f" • {key.replace('_', ' ').title()}: {value}")
        print("\n LECCIÓN CLAVE:")
        print(" La comprensión del negocio no es 'tomar notas'; es traducir necesidades vagas")
        print(" en objetivos analíticos específicos, medibles y alineados con restricciones reales.")

    def fase_2_comprension_datos(self):
        """Simula la fase de comprensión de datos: exploración inicial y evaluación de idoneidad"""
        print("\n" + "="*70)
        print(" FASE 2: COMPRENSIÓN DE LOS DATOS")
        print("="*70)
        # Generar datos sintéticos realistas
        print("\nGenerando datos transaccionales sintéticos...")
        self.generar_datos_sinteticos()
        # Exploración inicial
        print(f"\nExploración inicial de los datos:")
        print(f" • Total de transacciones: {len(self.datos_transaccionales):,}")
        print(f" • Periodo: {self.datos_transaccionales['fecha'].min().strftime('%Y-%m-%d')} a {self.datos_transaccionales['fecha'].max().strftime('%Y-%m-%d')}")
        print(f" • Categorías únicas: {self.datos_transaccionales['categoria'].nunique()}")
        print(f" • Productos únicos: {self.datos_transaccionales['producto_id'].nunique():,}")
        # Evaluación de idoneidad
        print(f"\nEvaluación de idoneidad para el objetivo analítico:")
        print(f" • ¿Contiene información de categoría de producto? Sí")
        print(f" • ¿Contiene información de valor de transacción? Sí")
        print(f" • ¿Cubre suficiente periodo para detectar patrones? 12 meses")
        print(f" • ¿Incluye datos demográficos del cliente? No (limitación identificada)")
        print(f"\n LIMITACIONES IDENTIFICADAS:")
        print(f" • Sin datos demográficos: no se pueden segmentar asociaciones por perfil de cliente")
        print(f" • Sin datos de inventario: no se puede verificar disponibilidad de productos asociados")
        print(f" → Decisión: Proceder con análisis a nivel de categoría/producto, no cliente")

    def generar_datos_sinteticos(self, transacciones=100000):
        """Genera datos transaccionales sintéticos seguros y realistas"""
        categorias = [
            "Alimentos", "Bebidas", "Lácteos", "Panadería",
            "Electrónica", "Hogar", "Ropa", "Juguetes"
        ]
        subcategorias = {
            "Alimentos": ["Carnes", "Verduras", "Frutas", "Congelados"],
            "Bebidas": ["Agua", "Refrescos", "Cerveza", "Vino"],
            "Lácteos": ["Leche", "Queso", "Yogur", "Mantequilla"],
            "Panadería": ["Pan", "Pasteles", "Galletas", "Facturas"],
            "Electrónica": ["Audio", "Video", "Computación", "Celulares"],
            "Hogar": ["Limpieza", "Decoración", "Cocina", "Baño"],
            "Ropa": ["Hombre", "Mujer", "Niños", "Accesorios"],
            "Juguetes": ["Educación", "Acción", "Muñecas", "Juegos"]
        }
        transacciones_lista = []
        fecha_base = datetime(2023, 1, 1)
        for i in range(transacciones):
            categoria = random.choice(categorias)
            subcategoria = random.choice(subcategorias[categoria])
            producto_id = f"{categoria[:3].upper()}{subcategoria[:3].upper()}{i%1000:03d}"
            # Precios realistas por categoría
            precio_base = {
                "Alimentos": 15.0, "Bebidas": 8.0, "Lácteos": 12.0, "Panadería": 10.0,
                "Electrónica": 250.0, "Hogar": 45.0, "Ropa": 85.0, "Juguetes": 35.0
            }
            precio = precio_base[categoria] * random.uniform(0.8, 1.2)
            # Cantidades realistas
            cantidad = random.randint(1, 5) if categoria != "Electrónica" else 1
            # Fecha con estacionalidad
            dias = random.randint(0, 364)
            fecha = fecha_base + timedelta(days=dias)
            transacciones_lista.append({
                "transaccion_id": i + 1,
                "fecha": fecha,
                "categoria": categoria,
                "subcategoria": subcategoria,
                "producto_id": producto_id,
                "cantidad": cantidad,
                "precio_unitario": round(precio, 2),
                "total": round(cantidad * precio, 2)
            })
        self.datos_transaccionales = pd.DataFrame(transacciones_lista)

    def fase_3_preparacion_datos(self):
        """Simula la fase de preparación de datos: limpieza, transformación y enriquecimiento"""
        print("\n" + "="*70)
        print(" FASE 3: PREPARACIÓN DE LOS DATOS")
        print("="*70)
        datos_originales = len(self.datos_transaccionales)
        # Limpieza de datos
        print("\n1. LIMPIEZA DE DATOS:")
        # Eliminar transacciones de prueba (total = 0)
        mask_prueba = self.datos_transaccionales["total"] == 0
        transacciones_prueba = mask_prueba.sum()
        if transacciones_prueba > 0:
            self.datos_transaccionales = self.datos_transaccionales[~mask_prueba]
            print(f" • Eliminadas {transacciones_prueba:,} transacciones de prueba (total=0)")
        # Corregir precios negativos (errores de sistema)
        mask_negativos = self.datos_transaccionales["precio_unitario"] < 0
        precios_negativos = mask_negativos.sum()
        if precios_negativos > 0:
            self.datos_transaccionales.loc[mask_negativos, "precio_unitario"] = abs(self.datos_transaccionales.loc[mask_negativos, "precio_unitario"])
            print(f" • Corregidos {precios_negativos:,} precios negativos")
        # Manejo de valores nulos
        nulos = self.datos_transaccionales.isnull().sum().sum()
        if nulos > 0:
            print(f" • Eliminadas {nulos:,} filas con valores nulos")
            self.datos_transaccionales = self.datos_transaccionales.dropna()
        # Transformación y enriquecimiento
        print("\n2. TRANSFORMACIÓN Y ENRIQUECIMIENTO:")
        # Agregar atributos derivados
        self.datos_transaccionales["mes"] = self.datos_transaccionales["fecha"].dt.month
        self.datos_transaccionales["dia_semana"] = self.datos_transaccionales["fecha"].dt.dayofweek
        self.datos_transaccionales["ticket_alto"] = self.datos_transaccionales["total"] > 100
        # Crear atributo de complementariedad (simulando conocimiento de dominio)
        complementariedad = {
            ("Vino", "Queso"): "alta",
            ("Pan", "Mantequilla"): "alta",
            ("Cerveza", "Carnes"): "alta",
            ("Electrónica", "Accesorios"): "alta"
        }
        self.datos_transaccionales["complementariedad"] = "baja"
        for (cat1, cat2), nivel in complementariedad.items():
            mask = (
                ((self.datos_transaccionales["subcategoria"] == cat1) &
                 (self.datos_transaccionales["subcategoria"].shift(-1) == cat2)) |
                ((self.datos_transaccionales["subcategoria"] == cat2) &
                 (self.datos_transaccionales["subcategoria"].shift(-1) == cat1))
            )
            self.datos_transaccionales.loc[mask, "complementariedad"] = nivel
        print(" • Agregado mes y día de la semana")
        print(" • Creado indicador de ticket alto (> $100)")
        print(" • Enriquecido con atributo de complementariedad basado en conocimiento de dominio")
        # Validación final
        datos_finales = len(self.datos_transaccionales)
        print(f"\nVALIDACIÓN FINAL:")
        print(f" • Datos originales: {datos_originales:,}")
        print(f" • Datos después de preparación: {datos_finales:,}")
        print(f" • Tasa de retención: {datos_finales/datos_originales*100:.1f}%")
        print(f" • Listos para análisis exploratorio y modelado")

    def demostrar_valor_kdd(self):
        """Demuestra el valor del proceso KDD estructurado vs. enfoque ad-hoc"""
        print("\n" + "="*70)
        print(" VALOR DEL PROCESO KDD: Estructurado vs. Ad-hoc")
        print("="*70)
        print("\n[ENFOQUE AD-HOC] Sin proceso KDD:")
        print(" • Aplicar algoritmo Apriori a todos los datos sin preparación")
        print(" • Resultado: Patrones obvios ('pan y leche') sin valor accionable")
        print(" • Tiempo invertido: 2 días | Valor generado: $0")
        print("\n[ENFOQUE KDD] Con proceso estructurado:")
        print(" • Fase 1: Definir objetivo claro (aumentar ticket en 15%)")
        print(" • Fase 2: Evaluar datos y identificar limitaciones (sin datos demográficos)")
        print(" • Fase 3: Preparar datos con enriquecimiento (complementariedad, ticket alto)")
        print(" • Resultado: Patrones accionables ('vinos y quesos en fines de semana')")
        print(" • Tiempo invertido: 5 días | Valor generado: Campaña que aumenta ticket en 18%")
        print("\n IMPACTO CUANTIFICABLE:")
        print(" • Inversión adicional en fases iniciales: 3 días")
        print(" • Retorno: $45,000 en ingresos adicionales (estimado para tienda mediana)")
        print(" • ROI de las fases iniciales del KDD: ($45,000 - $0) / (3 días × $500/día) = 3,000%")
        print("\n LECCIÓN FUNDAMENTAL:")
        print(" Las fases iniciales del KDD (comprensión y preparación) no son 'gastos';")
        print(" son inversiones que multiplican el valor del modelado posterior.")
        print(" Saltarlas garantiza hallazgos técnicamente correctos pero comercialmente irrelevantes.")
```

## Ejecución del simulador original
En esta sección se ejecuta la versión original del simulador, respetando las fases iniciales
mostradas en el ejemplo base del PDF.

```python
kdd = SimuladorKDDRetail()
kdd.fase_1_comprension_negocio()
kdd.fase_2_comprension_datos()
kdd.fase_3_preparacion_datos()
kdd.demostrar_valor_kdd()
```

## Modificaciones realizadas

Con base en el apartado "Tu turno", se desarrolló una versión mejorada del simulador para:
1. incluir una fase de evaluación,
2. diseñar métricas específicas para validar si los patrones aumentan el valor del ticket,
3. implementar retroalimentación al negocio y cerrar el ciclo del proceso KDD.

```python
class SimuladorKDDRetailMejorado(SimuladorKDDRetail):
    def fase_5_evaluacion(self):
        """Simula la fase 5 del KDD: Evaluación de patrones mediante A/B Testing"""
        print("\n" + "="*70)
        print(" FASE 5: EVALUACIÓN DE PATRONES")
        print("="*70)
        
        print("\n[MÉTRICAS ESPECÍFICAS] Entorno de prueba controlado (A/B Testing):")
        print(" • Grupo Control: Clientes bajo el esquema tradicional.")
        print(" • Grupo Piloto: Clientes expuestos a sugerencias de alta complementariedad (ej. 'Vino y Queso').")
        
        # Simulación de resultados
        ticket_promedio_historico = self.datos_transaccionales['total'].mean()
        incremento_simulado = random.uniform(0.12, 0.18) # Aumento entre 12% y 18%
        ticket_promedio_piloto = ticket_promedio_historico * (1 + incremento_simulado)
        
        print(f"\n Resultados del Piloto (1 mes de prueba):")
        print(f" • Ticket Promedio Grupo Control: ${ticket_promedio_historico:.2f}")
        print(f" • Ticket Promedio Grupo Piloto: ${ticket_promedio_piloto:.2f}")
        print(f" • Incremento Logrado: {incremento_simulado*100:.1f}%")
        
        # Validación contra el objetivo de la Fase 1
        objetivo_esperado = 15.0 
        print("\n[VALIDACIÓN DE RESULTADOS]:")
        if incremento_simulado * 100 >= objetivo_esperado:
            print(" ✅ ÉXITO: El incremento superó el objetivo del 15%. Los patrones descubiertos son valiosos.")
        else:
            print(" ⚠️ ALERTA: El incremento fue positivo pero no alcanzó el 15%. Se requiere iterar el modelo.")

    def retroalimentacion_negocio(self):
        """Mecanismo para cerrar el ciclo del proceso KDD"""
        print("\n" + "="*70)
        print(" RETROALIMENTACIÓN AL NEGOCIO Y CIERRE DE CICLO")
        print("="*70)
        print(" 1. Generación de Dashboard automatizado para los gerentes de tienda con las asociaciones descubiertas.")
        print(" 2. Actualización de reglas de negocio en el sistema de Punto de Venta (POS) basado en el piloto.")
        print(" 3. Siguiente iteración: Recolectar datos demográficos (limitación detectada en Fase 2) para afinar la campaña.")
```

## Ejecución del simulador mejorado

En esta sección se ejecuta la versión extendida del simulador, incorporando evaluación
y retroalimentación al negocio.

```python
# Instanciar el simulador mejorado
kdd_mejorado = SimuladorKDDRetailMejorado()

# Ejecutar pipeline
kdd_mejorado.fase_1_comprension_negocio()
kdd_mejorado.fase_2_comprension_datos()
kdd_mejorado.fase_3_preparacion_datos()
kdd_mejorado.demostrar_valor_kdd()

# Ejecutar las nuevas fases
kdd_mejorado.fase_5_evaluacion()
kdd_mejorado.retroalimentacion_negocio()
```

## Glosario

**KDD:** Proceso sistemático de descubrimiento de conocimiento en bases de datos.  
**Comprensión del negocio:** Fase donde se traduce un problema general en un objetivo analítico.  
**Preparación de datos:** Limpieza, transformación y enriquecimiento de la información.  
**Evaluación:** Validación de los resultados obtenidos respecto al objetivo del negocio.  
**Retroalimentación:** Comunicación de hallazgos para apoyar la toma de decisiones y reiniciar el ciclo analítico.
